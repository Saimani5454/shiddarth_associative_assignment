## Define Category and Sub-Category Logic

### Subtask:
Develop functions or conditional logic to categorize items based on keywords found in the 'GOODS DESCRIPTION' column and 'HS CODE'. This will involve creating a 'Category' column (e.g., Electronics, Glass, Steel, Plastic, Polyhouse) and a 'Sub-Category' column (e.g., Borosilicate, Opalware, Spoon, Fork) based on the assigned Category.


**Reasoning**:
I need to define a function `categorize_goods` that takes `goods_desc` and `hs_code` as input and applies conditional logic to assign 'Category' and 'Sub-Category' based on keywords in `goods_desc` and specific `hs_code` values, as per the instructions. This function will then be applied to the DataFrame to create the new columns.



```python
def categorize_goods(goods_desc, hs_code):
    goods_desc_lower = str(goods_desc).lower()
    category = 'Other'
    sub_category = None

    # Category: Glass
    if 'glass' in goods_desc_lower:
        category = 'Glass'
        if 'borosilicate' in goods_desc_lower:
            sub_category = 'Borosilicate'
        elif 'opalware' in goods_desc_lower:
            sub_category = 'Opalware'
        elif 'tumbler' in goods_desc_lower:
            sub_category = 'Tumbler'
        elif 'jug' in goods_desc_lower:
            sub_category = 'Jug'
        elif 'bowl' in goods_desc_lower:
            sub_category = 'Bowl'
        elif 'plate' in goods_desc_lower:
            sub_category = 'Plate'

    # Category: Steel
    elif 'steel' in goods_desc_lower or 'cutlery' in goods_desc_lower:
        category = 'Steel'
        if 'spoon' in goods_desc_lower:
            sub_category = 'Spoon'
        elif 'fork' in goods_desc_lower:
            sub_category = 'Fork'
        elif 'knife' in goods_desc_lower:
            sub_category = 'Knife'
        elif 'holder' in goods_desc_lower:
            sub_category = 'Cutlery Holder'
        elif 'scrubber' in goods_desc_lower:
            sub_category = 'Scrubber'
        elif 'strainer' in goods_desc_lower:
            sub_category = 'Strainer'
        elif 'kettle' in goods_desc_lower:
            sub_category = 'Kettle'
        elif 'cooker' in goods_desc_lower:
            sub_category = 'Cooker'
        elif 'frying pan' in goods_desc_lower:
            sub_category = 'Frying Pan'
        elif 'tava' in goods_desc_lower:
            sub_category = 'Tava'
        elif 'bottle' in goods_desc_lower:
            sub_category = 'Bottle'

    # Category: Plastic
    elif 'plastic' in goods_desc_lower or 'melamine' in goods_desc_lower:
        category = 'Plastic'
        if 'bottle' in goods_desc_lower:
            sub_category = 'Bottle'
        elif 'container' in goods_desc_lower:
            sub_category = 'Container'
        elif 'tray' in goods_desc_lower:
            sub_category = 'Tray'
        elif 'mug' in goods_desc_lower:
            sub_category = 'Mug'

    # Category: Polyhouse/Greenhouse
    elif 'polyhouse' in goods_desc_lower or 'greenhouse' in goods_desc_lower:
        category = 'Polyhouse'
        if 'film' in goods_desc_lower:
            sub_category = 'Film'
        elif 'pipe' in goods_desc_lower:
            sub_category = 'Pipe'
        elif 'sheet' in goods_desc_lower:
            sub_category = 'Sheet'

    # Category: Electronics (Example using HS Code, adjust as needed)
    # These are illustrative HS Codes; real ones would need to be verified.
    elif hs_code in [85161000, 85167990, 85168000]: # Example HS codes for heating appliances, other electrical appliances
        category = 'Electronics'
        if 'heater' in goods_desc_lower:
            sub_category = 'Heater'
        elif 'mixer' in goods_desc_lower:
            sub_category = 'Mixer'
        elif 'iron' in goods_desc_lower:
            sub_category = 'Iron'

    # General Kitchenware category based on common HS Codes if not caught by material
    if category == 'Other' and str(hs_code).startswith('7323'): # HS Code for Table, kitchen or other household articles
        category = 'Kitchenware'
        if 'cooker' in goods_desc_lower:
            sub_category = 'Pressure Cooker'
        elif 'utensil' in goods_desc_lower:
            sub_category = 'Utensil'
        elif 'frypan' in goods_desc_lower or 'frying pan' in goods_desc_lower:
            sub_category = 'Frying Pan'
        elif 'kadai' in goods_desc_lower:
            sub_category = 'Kadai'
        elif 'casserole' in goods_desc_lower:
            sub_category = 'Casserole'
        elif 'pot' in goods_desc_lower:
            sub_category = 'Pot'
        elif 'lid' in goods_desc_lower:
            sub_category = 'Lid'
        elif 'grill' in goods_desc_lower:
            sub_category = 'Grill'

    return category, sub_category

# Apply the function to create 'Category' and 'Sub-Category' columns
df[['Category', 'Sub-Category']] = df.apply(lambda row: categorize_goods(row['GOODS DESCRIPTION'], row['HS CODE']), axis=1, result_type='expand')

print("Created 'Category' and 'Sub-Category' columns.")
df.head()
```

```python
query = """WITH yearly AS (
    SELECT
        strftime('%Y', DATE) AS year,
        SUM("TOTAL VALUE_INR") AS total_value,
        SUM("DUTY PAID_INR") AS total_duty,
        SUM("Grand Total") AS total_grand
    FROM trade_data
    GROUP BY strftime('%Y', DATE)
)
SELECT
    a.year,
    a.total_value,
    b.total_value AS prev_year_value,
    ROUND(((a.total_value - b.total_value) / b.total_value) * 100.0, 2) AS YoY_Total_Value_Growth,

    a.total_duty,
    b.total_duty AS prev_year_duty,
    ROUND(((a.total_duty - b.total_duty) / b.total_duty) * 100.0, 2) AS YoY_Duty_Growth,

    a.total_grand,
    b.total_grand AS prev_year_grand,
    ROUND(((a.total_grand - b.total_grand) / b.total_grand) * 100.0, 2) AS YoY_Grand_Total_Growth

FROM yearly a
LEFT JOIN yearly b
ON CAST(a.year AS INTEGER) = CAST(b.year AS INTEGER) + 1
ORDER BY a.year;
"""
pd.read_sql(query, conn)
```

```python
query="""WITH ranked AS (
    SELECT
        "HS CODE" AS hsn_code,
        SUM("Grand Total") AS total_value,
        ROW_NUMBER() OVER (ORDER BY SUM("Grand Total") DESC) AS rn
    FROM trade_data
    GROUP BY "HS CODE"
),
bucketed AS (
    SELECT
        CASE WHEN rn <= 25 THEN hsn_code ELSE 'OTHERS' END AS category,
        total_value
    FROM ranked
)
SELECT
    category,
    SUM(total_value) AS total_value,
    ROUND(SUM(total_value) * 100.0 / (SELECT SUM("Grand Total") FROM trade_data), 2) AS contribution_percent
FROM bucketed
GROUP BY category
ORDER BY total_value DESC;
"""
pd.read_sql(query, conn)
```

```python
# Retry: robust version that avoids None columns and handles missing fields gracefully.
import pandas as pd
import numpy as np
import os
from scipy import stats
# Removed: from caas_jupyter_tools import display_dataframe_to_user

INPUT = "/content/processed_trade_data.csv"
# df = pd.read_csv(INPUT) # Removed this line to ensure we use the 'df' with Grand_Total and categories

# normalize column names
df.columns = [c.strip() for c in df.columns]

def find_col(possible_names):
    for p in possible_names:
        for c in df.columns:
            if c.lower().replace('_',' ').strip() == p.lower().replace('_',' ').strip():
                return c
    for p in possible_names:
        for c in df.columns:
            if p.lower() in c.lower():
                return c
    return None

col_shipment_date = find_col(['shipment_date','date_of_shipment','date'])
col_year = find_col(['year'])
col_hsn = find_col(['hsn_code','hsn'])
col_model = find_col(['model','model_name','model_name_extracted','model_name_extracted'])
col_model_number = find_col(['model_number','model_number_extracted'])
col_capacity = find_col(['parsed_capacity','capacity','capacity/spec','capacity_spec','parsed capacity','capacity/spec'])
col_material = find_col(['parsed_material','material','material_type'])
col_embedded_qty = find_col(['parsed_embedded_qty','embedded_quantity','embedded_qty'])
col_qty = find_col(['quantity','qty','quantity '])
col_unit = find_col(['unit_standard','unit','standardized_unit'])
col_unit_price_inr = find_col(['unit price (inr)','unit_price_inr','unit_price_inr','unit price inr','unit price'])
col_unit_price_usd = find_col(['parsed_price_usd','unit_price_usd','unit price usd','price_usd'])
col_total_value = find_col(['total value (inr)','total_value_inr','total value inr','total_value','total_value'])
col_duty = find_col(['duty paid (inr)','duty_paid','duty paid','duty_paid_inr','duty'])
col_grand = find_col(['grand total (inr)','grand_total','grand total','grand_total_inr','grand'])
col_duty_pct = find_col(['duty_pct_of_value','duty_pct_of_value','duty_pct','duty_pct'])
col_supplier = find_col(['supplier name and address','supplier','supplier_name_and_address','supplier_name','supplier_name'])

# Convert numeric columns carefully
for c in [col_qty, col_unit_price_inr, col_unit_price_usd, col_total_value, col_duty, col_grand, col_embedded_qty]:
    if c and c in df.columns:
        df[c] = pd.to_numeric(df[c].astype(str).str.replace(',','').replace('', np.nan), errors='coerce')

# ensure year column
if col_year is None:
    if col_shipment_date and col_shipment_date in df.columns:
        df['year'] = pd.to_datetime(df[col_shipment_date], dayfirst=True, errors='coerce').dt.year
        col_year = 'year'
    else:
        df['year'] = pd.NA
        col_year = 'year'

# Model-Level Insights
model_col = col_model
year_col = col_year
qty_col = col_qty
# prefer INR unit price, else USD
unit_price_col = col_unit_price_inr if col_unit_price_inr else col_unit_price_usd

# Build aggregation dict dynamically
if model_col and year_col and qty_col and model_col in df.columns:
    agg_dict = {}
    agg_dict[qty_col] = 'sum'
    if unit_price_col and unit_price_col in df.columns:
        agg_dict[unit_price_col] = ['mean','min','max','median','count']
    if col_supplier and col_supplier in df.columns:
        agg_dict[col_supplier] = lambda x: x.nunique()
    grouped = df.groupby([model_col, year_col], dropna=False).agg(agg_dict)
    # flatten columns
    grouped.columns = ['_'.join([c for c in tup if c]) if isinstance(tup, tuple) else tup for tup in grouped.columns.values]
    grouped = grouped.reset_index().rename(columns={qty_col+'_'+'sum':'total_qty'})
    # rename supplier lambda column if present
    if col_supplier and (col_supplier in df.columns):
        # find column name containing supplier lambda
        for col in grouped.columns:
            if 'lambda' in str(col):
                grouped = grouped.rename(columns={col:'supplier_count'})
    # standardize price column names if present
    if unit_price_col and unit_price_col in df.columns:
        # mean is unit_price_col_mean
        mean_col = unit_price_col + '_mean'
        min_col = unit_price_col + '_min'
        max_col = unit_price_col + '_max'
        med_col = unit_price_col + '_median'
        count_col = unit_price_col + '_count'
        # rename to friendlier names if exist
        rename_map = {}
        if mean_col in grouped.columns:
            rename_map[mean_col] = 'avg_unit_price'
        if min_col in grouped.columns:
            rename_map[min_col] = 'min_unit_price'
        if max_col in grouped.columns:
            rename_map[max_col] = 'max_unit_price'
        if med_col in grouped.columns:
            rename_map[med_col] = 'median_unit_price'
        if count_col in grouped.columns:
            rename_map[count_col] = 'price_observations'
        grouped = grouped.rename(columns=rename_map)
else:
    grouped = pd.DataFrame()

# Save model-year summary
model_year_csv = "/content/model_year_summary.csv"
grouped.to_csv(model_year_csv, index=False)

# Compare Unit Prices for same model across suppliers
if model_col and col_supplier and unit_price_col and all(x in df.columns for x in [model_col, col_supplier, unit_price_col]):
    comp = df.groupby([model_col, col_supplier]).agg(
        avg_price = (unit_price_col, 'mean'),
        min_price = (unit_price_col, 'min'),
        max_price = (unit_price_col, 'max'),
        total_qty = (qty_col, 'sum'),
        shipments = (unit_price_col, 'count')
    ).reset_index().sort_values([model_col, 'avg_price'])
else:
    comp = pd.DataFrame()

comp_csv = "/content/model_supplier_prices.csv"
comp.to_csv(comp_csv, index=False)

# Capacity Analysis
capacity_col = col_capacity
if capacity_col and capacity_col in df.columns:
    cap = df.groupby(capacity_col).agg(
        total_qty = (qty_col, 'sum'),
        total_value = (col_total_value, 'sum'),
        shipments = (capacity_col, 'count')
    ).reset_index().sort_values('total_qty', ascending=False)
else:
    cap = pd.DataFrame()

cap_csv = "/content/capacity_summary.csv"
cap.to_csv(cap_csv, index=False)

# Cost & Duty Structure
df['quantity_for_landed'] = df[qty_col].replace(0, np.nan) if qty_col and qty_col in df.columns else np.nan
if col_grand and col_grand in df.columns and 'quantity_for_landed' in df.columns:
    df['landed_cost_per_unit'] = df[col_grand] / df['quantity_for_landed']
else:
    df['landed_cost_per_unit'] = np.nan

# duty percent
if col_duty_pct and col_duty_pct in df.columns:
    df['duty_pct'] = pd.to_numeric(df[col_duty_pct], errors='coerce')
elif col_duty and col_total_value and col_duty in df.columns and col_total_value in df.columns:
    df['duty_pct'] = df[col_duty] / df[col_total_value] * 100
else:
    df['duty_pct'] = np.nan

# Flag anomalies: duty_pct < 0 or > 100 or extreme z-score
duty_pct_series = df['duty_pct'].dropna()
anomaly_flags = pd.Series(False, index=df.index)
if not duty_pct_series.empty:
    # compute zscore on non-null entries
    z = stats.zscore(duty_pct_series)
    z_series = pd.Series(z, index=duty_pct_series.index)
    full_z = pd.Series(index=df.index, dtype=float)
    full_z.loc[z_series.index] = z_series.values
    anomaly_flags = (df['duty_pct'] < 0) | (df['duty_pct'] > 100) | (full_z.abs() > 3)
else:
    anomaly_flags[:] = False

df['anomalous_duty_flag'] = anomaly_flags

anomalies = df[df['anomalous_duty_flag']].copy()
anomalies_csv = "/content/anomalous_duty_transactions.csv"
anomalies.to_csv(anomalies_csv, index=False)

# Save enriched dataset
enriched_csv = "/content/processed_trade_data_enriched.csv"
df.to_csv(enriched_csv, index=False)

# Display outputs
print("Model-Year Summary (Top 50 rows)")
if not grouped.empty:
    print(grouped.head(50).to_markdown(index=False))
else:
    print("Empty DataFrame")

print("\nModel-Supplier Price Comparison (Top 50 rows)")
if not comp.empty:
    print(comp.head(50).to_markdown(index=False))
else:
    print("Empty DataFrame")

print("\nTop Capacities by Volume (Top 50 rows)")
if not cap.empty:
    print(cap.head(50).to_markdown(index=False))
else:
    print("Empty DataFrame")

print("\nAnomalous Duty Transactions (Top 50 rows)")
if not anomalies.empty:
    print(anomalies.head(50).to_markdown(index=False))
else:
    print("Empty DataFrame")

print("\nSaved CSVs:")
print(" - Model-Year summary:", model_year_csv)
print(" - Model-Supplier prices:", comp_csv)
print(" - Capacity summary:", cap_csv)
print(" - Anomalous transactions:", anomalies_csv)
print(" - Full enriched dataset:", enriched_csv)
```

```python
import sqlite3
import pandas as pd

# Re-establish connection (if not already connected or if session restarted)
conn = sqlite3.connect("trade_database.db")

# Explicitly load the fully enriched DataFrame from the CSV file
# This ensures all calculated columns (like Grand_Total, Category, Sub-Category) are present
df = pd.read_csv("/content/processed_trade_data_enriched.csv")

# Ensure 'Grand_Total' is present in the DataFrame before saving to SQL
# This step is added as a safeguard if previous steps failed to include it in the CSV
if 'Grand_Total' not in df.columns:
    df['Grand_Total'] = df['TOTAL VALUE_INR'] + df['DUTY PAID_INR']

# Assume 'df' is the current DataFrame in memory, containing 'Grand_Total'
# and other enriched columns. Re-saving it to ensure the database table is up-to-date.
# This step ensures 'Grand_Total' is available in the 'processed_trade_data_enriched' table.
print("Columns in DataFrame before saving to SQL:", df.columns.tolist())
df.to_sql("processed_trade_data_enriched", conn, if_exists="replace", index=False)

query="""SELECT
    Model AS model,
    NULL AS supplier,
    "Capacity/Spec" AS capacity,
    QUANTITY AS quantity,
    Grand_Total AS total_cost,
    CAST(Grand_Total AS REAL) / NULLIF(CAST(QUANTITY AS REAL),0) AS raw_landed_cost_per_unit,
    ROUND(CAST(Grand_Total AS REAL) / NULLIF(CAST(QUANTITY AS REAL),0),2) AS landed_cost_per_unit
FROM processed_trade_data_enriched;
"""
pd.read_sql(query,conn)
```

Identify Anomalies in Duty Structure

Conditions:

Duty % < 0

Duty % > 100

Duty % > (Mean + 2 Std Dev)

```python
# Fetch the entire enriched dataset into a pandas DataFrame
df_enriched = pd.read_sql("SELECT * FROM processed_trade_data_enriched", conn)

# Calculate mean and standard deviation of duty_pct using pandas
duty_pct_series = df_enriched['duty_pct'].dropna()

if not duty_pct_series.empty:
    avg_duty = duty_pct_series.mean()
    std_duty = duty_pct_series.std()
    # Define the upper bound for statistical outliers (mean + 2 * std_dev)
    outlier_threshold = avg_duty + 2 * std_duty
else:
    avg_duty = 0
    std_duty = 0
    outlier_threshold = float('inf') # No outliers if series is empty

# Apply anomaly flagging logic using pandas
df_flagged = df_enriched.copy()
df_flagged['anomaly_flag'] = 'NORMAL'

df_flagged.loc[df_flagged['duty_pct'] < 0, 'anomaly_flag'] = 'NEGATIVE DUTY ERROR'
df_flagged.loc[df_flagged['duty_pct'] > 100, 'anomaly_flag'] = 'UNREALISTIC (>100%)'
# Only flag as statistical outlier if not already flagged by other conditions
df_flagged.loc[
    (df_flagged['duty_pct'] > outlier_threshold) &
    (df_flagged['anomaly_flag'] == 'NORMAL'),
    'anomaly_flag'
] = 'STATISTICAL OUTLIER'

# Select and display only the anomalous transactions
anomalous_transactions = df_flagged[df_flagged['anomaly_flag'] != 'NORMAL'].sort_values(by='duty_pct', ascending=False)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

print("Anomalous Duty Transactions (flagged by Pandas calculations):")
if not anomalous_transactions.empty:
    print(anomalous_transactions.to_markdown(index=False))
else:
    print("No anomalous duty transactions found.")
```

```python
all_queries_content = []

# Query from cell kivCRZIPi8Sc (Yearly Growth Analysis)
yearly_growth_query = """WITH yearly AS (
    SELECT
        strftime('%Y', DATE) AS year,
        SUM("TOTAL VALUE_INR") AS total_value,
        SUM("DUTY PAID_INR") AS total_duty,
        SUM("Grand Total") AS total_grand
    FROM trade_data
    GROUP BY strftime('%Y', DATE)
)
SELECT
    a.year,
    a.total_value,
    b.total_value AS prev_year_value,
    ROUND(((a.total_value - b.total_value) / b.total_value) * 100.0, 2) AS YoY_Total_Value_Growth,

    a.total_duty,
    b.total_duty AS prev_year_duty,
    ROUND(((a.total_duty - b.total_duty) / b.total_duty) * 100.0, 2) AS YoY_Duty_Growth,

    a.total_grand,
    b.total_grand AS prev_year_grand,
    ROUND(((a.total_grand - b.total_grand) / b.total_grand) * 100.0, 2) AS YoY_Grand_Total_Growth

FROM yearly a
LEFT JOIN yearly b
ON CAST(a.year AS INTEGER) = CAST(b.year AS INTEGER) + 1
ORDER BY a.year;
"""
all_queries_content.append("-- Yearly Growth Analysis Query\n" + yearly_growth_query)

# Query from cell dPM1L5NSi9AR (HSN Code Contribution)
hsn_contribution_query = """WITH ranked AS (
    SELECT
        "HS CODE" AS hsn_code,
        SUM("Grand Total") AS total_value,
        ROW_NUMBER() OVER (ORDER BY SUM("Grand Total") DESC) AS rn
    FROM trade_data
    GROUP BY "HS CODE"
),
bucketed AS (
    SELECT
        CASE WHEN rn <= 25 THEN hsn_code ELSE 'OTHERS' END AS category,
        total_value
    FROM ranked
)
SELECT
    category,
    SUM(total_value) AS total_value,
    ROUND(SUM(total_value) * 100.0 / (SELECT SUM("Grand Total") FROM trade_data), 2) AS contribution_percent
FROM bucketed
GROUP BY category
ORDER BY total_value DESC;
"""
all_queries_content.append("-- HSN Code Contribution Query\n" + hsn_contribution_query)

# Query from cell jwWHk1hHjsIp (Entity Activity)
entity_activity_query = """WITH activity AS (
    SELECT
        IEC AS entity_identifier,
        MIN(strftime('%Y', DATE)) AS first_year,
        MAX(strftime('%Y', DATE)) AS last_year
    FROM trade_data
    GROUP BY IEC
)
SELECT
    entity_identifier,
    first_year,
    last_year,
    CASE
        WHEN last_year = '2025' THEN 'ACTIVE'
        ELSE 'CHURNED'
    END AS entity_status
FROM activity
ORDER BY entity_status, entity_identifier;
"""
all_queries_content.append("-- Entity Activity Query\n" + entity_activity_query)

# Query from cell nkqWztbsl33a (Model-Year Summary)
model_year_summary_query = """SELECT
    Model AS model,
    strftime('%Y', DATE) AS year,
    SUM(QUANTITY) AS total_quantity,
    AVG("UNIT PRICE_INR") AS avg_unit_price_inr,
    AVG("UNIT PRICE_USD") AS avg_unit_price_usd
FROM processed_trade_data_enriched
GROUP BY Model, strftime('%Y', DATE)
ORDER BY model, year;
"""
all_queries_content.append("-- Model-Year Summary Query\n" + model_year_summary_query)

# Query from cell dJ_PZkgUmp68 (Model-Specific Price Analysis)
model_price_analysis_query = """SELECT
    Model AS model,
    AVG("UNIT PRICE_INR") AS avg_price_inr,
    MIN("UNIT PRICE_INR") AS min_price_inr,
    MAX("UNIT PRICE_INR") AS max_price_inr,
    SUM(QUANTITY) AS total_quantity,
    COUNT(*) AS shipment_count
FROM processed_trade_data_enriched
WHERE Model IS NOT NULL
GROUP BY Model
ORDER BY model, avg_price_inr ASC;
"""
all_queries_content.append("-- Model-Specific Price Analysis Query\n" + model_price_analysis_query)

# Query from cell koKumtifnC_w (Capacity Analysis)
capacity_analysis_query = """SELECT
    "Capacity/Spec" AS capacity,
    SUM(QUANTITY) AS total_quantity,
    SUM("TOTAL VALUE_INR") AS total_import_value,
    ROUND((SUM("TOTAL VALUE_INR") * 100) / (SELECT SUM("TOTAL VALUE_INR") FROM processed_trade_data_enriched), 2) AS percentage_contribution
FROM processed_trade_data_enriched
WHERE "Capacity/Spec" IS NOT NULL
GROUP BY "Capacity/Spec"
ORDER BY total_quantity DESC;
"""
all_queries_content.append("-- Capacity Analysis Query\n" + capacity_analysis_query)

# Query from cell XJOJGwU-neTj (Landed Cost Per Unit)
landed_cost_query = """SELECT
    Model AS model,
    NULL AS supplier,
    "Capacity/Spec" AS capacity,
    QUANTITY AS quantity,
    Grand_Total AS total_cost,
    CAST(Grand_Total AS REAL) / NULLIF(CAST(QUANTITY AS REAL),0) AS raw_landed_cost_per_unit,
    ROUND(CAST(Grand_Total AS REAL) / NULLIF(CAST(QUANTITY AS REAL),0),2) AS landed_cost_per_unit
FROM processed_trade_data_enriched;
"""
all_queries_content.append("-- Landed Cost Per Unit Query\n" + landed_cost_query)

# Query from cell GlBCCY_6pQmo (Trade Model Economics View)
trade_model_view_query = """DROP VIEW IF EXISTS trade_model_economics;
CREATE VIEW trade_model_economics AS
SELECT
    DATE AS shipment_date,
    strftime('%Y', DATE) AS year,
    Model AS model,
    "Capacity/Spec" AS capacity,
    NULL AS supplier,
    QUANTITY AS quantity,
    "Standardized Unit" AS unit,
    "UNIT PRICE_INR" AS unit_price_inr,
    "UNIT PRICE_USD" AS parsed_price_usd,
    Grand_Total AS grand_total,
    ROUND(CAST(Grand_Total AS REAL) / NULLIF(CAST(QUANTITY AS REAL),0), 2) AS landed_cost_per_unit,
    duty_pct
FROM processed_trade_data_enriched;
"""
all_queries_content.append("-- Trade Model Economics View Query\n" + trade_model_view_query)

# Write all queries to a single file
output_file_path = "/content/all_queries.sql"
with open(output_file_path, "w") as f:
    for query_block in all_queries_content:
        f.write(query_block)
        f.write("\n\n-- ==================================================\n\n") # Separator for readability

print(f"All SQL queries have been saved to: {output_file_path}")
```
